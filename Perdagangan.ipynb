**Data Jumlah Penduduk di Kalimantan Selatan**


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Mengirimkan permintaan GET ke halaman web
url = 'https://data.kalselprov.go.id/dataset/data/1123/bar'  # Ganti dengan URL yang sesuai
response = requests.get(url)

# Mengecek apakah permintaan berhasil
if response.status_code == 200:
    # Menginisialisasi BeautifulSoup untuk menguraikan halaman web
    soup = BeautifulSoup(response.text, 'html.parser')

    # Mencari tabel dengan class 'table-responsive'
    target_table = soup.find('div', {'class': 'table-responsive'})

    if target_table is not None:
        # Menemukan semua baris dalam tabel
        rows = target_table.find_all('tr')

        data = []
        # Mengambil data dari setiap baris
        for row in rows[:15]:
            # Mengambil semua sel dalam setiap baris
            cells = row.find_all(['th', 'td'])
            data.append([cell.get_text(strip=True) for cell in cells])

        # Membuat dataframe pandas
        df = pd.DataFrame(data[1:], columns=data[0])

        # Menampilkan dataframe pandas
        display(df)

    else:
        print("Tabel dengan class 'table-responsive' tidak ditemukan.")

else:
    print("Gagal mengambil halaman web.")
```

**Luas Wilayah di Kalimantan Selatan**

```python
import requests
from bs4 import BeautifulSoup
from tabulate import tabulate

# Tentukan URL yang akan di-scrape
url = 'https://id.wikipedia.org/wiki/Kalimantan_Selatan'

# Kirim GET request ke URL
response = requests.get(url)

# Periksa apakah permintaan berhasil
if response.status_code == 200:
    # Gunakan BeautifulSoup untuk mengambil konten halaman
    soup = BeautifulSoup(response.content, 'html.parser')

    # Cari tabel dengan class 'wikitable' pada halaman
    tables = soup.find_all('table', {'class': 'wikitable'})

    # Misalnya, jika Anda tertarik pada tabel ketiga, Anda dapat memilih indeks 2
    target_table = tables[2]

    # Inisialisasi list kosong untuk menyimpan data
    data = []

    # Looping untuk mengakses setiap baris dan kolom tabel
    for row in target_table.find_all('tr'):
        row_data = []
        cells = row.find_all(['th', 'td'])
        for i, cell in enumerate(cells):
            if i < 4:  # Menambahkan hanya empat kolom pertama
                row_data.append(cell.get_text(strip=True))
        if row_data:
            data.append(row_data)

    if len(data) > 1:
        data.pop(1)

    # Tampilkan tabel menggunakan tabulate
    print(tabulate(data, headers="firstrow", tablefmt="pipe"))
else:
    print("Failed to retrieve the webpage.")
```

```python
import matplotlib.pyplot as plt

# Data dari baris ke-14
data_provinsi_kalsel = qq.iloc[13]

# Menghapus kolom pertama (KODE wilayah)
data_provinsi_kalsel = data_provinsi_kalsel[1:]

# Membersihkan data dengan mengganti "-" menjadi 0
data_provinsi_kalsel_cleaned = data_provinsi_kalsel.replace('-', 0)

# Menggantikan NaN dengan 0
data_provinsi_kalsel_cleaned = data_provinsi_kalsel_cleaned.fillna(0)

# Mengganti tanda koma (",") dengan tanda titik (".") dan mengonversi ke tipe data float
data_provinsi_kalsel_cleaned = data_provinsi_kalsel_cleaned.str.replace(',', '', regex=True).astype(float)

# Menghapus nilai NaN
data_provinsi_kalsel_cleaned = data_provinsi_kalsel_cleaned.dropna()

# Mengonversi ke tipe data integer
data_provinsi_kalsel_cleaned = data_provinsi_kalsel_cleaned.astype(int)

# Nama kolom dan data untuk grafik bar chart
kategori = data_provinsi_kalsel_cleaned.index[1:]
jumlah = data_provinsi_kalsel_cleaned.values[1:]

# Mengurutkan data berdasarkan jumlah
sorted_data = sorted(zip(kategori, jumlah), key=lambda x: x[1], reverse=True)

# Memilih 15 kategori teratas
top_15_kategori = [x[0] for x in sorted_data[:15]]
top_15_jumlah = [x[1] for x in sorted_data[:15]]

# Membuat daftar warna dengan warna default
colors = ['lightgreen'] * len(top_15_kategori)

# Memberikan warna khusus pada batang ke-4 (indeks 3)
colors[4] = 'darkgreen'  # Merah untuk batang ke-4

# Membuat grafik bar chart dengan warna yang berbeda pada batang ke-4
plt.figure(figsize=(10, 6))
plt.barh(top_15_kategori, top_15_jumlah, color=colors)
plt.xlabel('Jumlah')
plt.title('15 Kategori Pekerjaan Teratas di Kalimantan Selatan')

# Menambahkan label angka pada batang grafik
for i, v in enumerate(top_15_jumlah):
    plt.text(v, i, str(v), ha='left', va='center')

plt.gca().invert_yaxis()  # Membalik urutan kategori agar yang terbesar ada di atas
plt.show()

```
